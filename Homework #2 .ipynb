```python
import pandas as pd
import numpy as np
import os 
import time
```

```python
# station data
st_filename=r'C:\Users\user\Desktop\iis_airbox_station.csv'
st_df=pd.read_csv(st_filename)
mask1=(st_df.create_year<=2018)
mask2=(st_df.station_name.str.contains('高雄'))
mask3=(st_df.station_address.str.contains('高雄'))
st_df_kaohsiung=st_df[mask1&mask2|mask3]
st_df_kaohsiung
# st_df.filter(like='2018',axis=0)
```

```python
# 1-5 airbox data
day_file=r'C:\Users\user\Desktop\iis_airbox_2020100{}.csv'
res=pd.read_csv(day_file.format(1))
for i in range(1,6):
    tmp_read=pd.read_csv(day_file.format(i))
    res=pd.concat([res,tmp_read],verify_integrity=True,join="inner", ignore_index = True)
del st_df_kaohsiung['station_name']
del st_df_kaohsiung['station_address']

res.rename(columns={"device_id":"station_id"},inplace = True)
res=pd.merge(res,st_df_kaohsiung,on='station_id',how='inner')

res
```

```python
feather_save_path=r"C:\Users\user\Desktop\df.feather"
json_save_path=r"C:\Users\user\Desktop\df.json"

start_time=time.time()
res.to_feather(feather_save_path)
end_time=time.time()
print(f'save file to feather during time: {end_time-start_time} sec')

start_time=time.time()
res.to_json(json_save_path)
end_time=time.time()
print(f'save file to json during time: {end_time-start_time} sec')


feather_size=os.path.getsize(feather_save_path)
json_size=os.path.getsize(json_save_path)
print(f'feather size: {feather_size} bytes \njson size: {json_size} bytes')
```
